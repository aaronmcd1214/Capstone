```python
from src import data_cleaning as dc
from src import fetch_data as fd
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
df_batting = pd.read_csv('src/data/baseballdatabank-master/core/Batting.csv')
df_fielding = pd.read_csv('src/data/baseballdatabank-master/core/Fielding.csv')
```

```python
df_batting = dc.initial_drop(df_batting)
```

```python
df_batting = dc.combine_stints(df_batting)
```

```python
df_batting = dc.map_position(df_batting, df_fielding)
```

```python
vc = pd.value_counts(df_batting['playerID'])
years_7 = df_batting[df_batting['playerID'].isin(vc[vc > 6].index)]
```

```python
years_7 = years_7.groupby('playerID').head(7)
pd.value_counts(years_7['playerID'])
```

```python
num_players = defaultdict(int)
num_players['7 years'] = len(years_7['playerID'].unique())
```

```python
num_players
```

```python
pos = ['C', '1B', '2B', '3B', 'SS', 'OF']

for p in pos:
    single_position = years_7[years_7['pos'] == p]
    key = '7yrs_' + p
    num_players[key] = len(single_position['playerID'].unique())

num_players
```

```python
ab = [400, 300, 200, 100]
pos = ['C', '1B', '2B', '3B', 'SS', 'OF']

for a in ab:
    for p in pos:
        pos_ab = years_7[(years_7['pos'] == p) & (years_7['AB'] >= a)]
        key = '7yrs_' + p + '_' + str(a)
        num_players[key] = len(pos_ab['playerID'].unique())
num_players
```

```python
for a in ab:
    pos_df = years_7[years_7['AB'] >= a]
    vc = pd.value_counts(pos_df['playerID'])
    pos_df = pos_df[pos_df['playerID'].isin(vc[vc == 7].index)]
    key = '7yrs_' + str(a)
    num_players[key] = len(pos_df['playerID'].unique())
num_players
```

```python
ab_400 = years_7[years_7['AB'] >= 100]
vc = pd.value_counts(ab_400['playerID'])
ab_400 = ab_400[ab_400['playerID'].isin(vc[vc == 7].index)]
len(ab_400['playerID'].unique())
```
