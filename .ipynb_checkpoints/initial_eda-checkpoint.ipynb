```python
#convert finalGame and debut to datetime objects
df_people['debut'] = pd.to_datetime(df_people['debut'])
df_people['finalGame'] = pd.to_datetime(df_people['finalGame'])
```

```python
# Active players = players whose last game was within the last two years
df_active_players = df_people[df_people['finalGame'] > pd.to_datetime('2015-01-01')]
```

```python
df_active_batters = pd.merge(df_batting, df_active_players, how='left', on='playerID')
```

```python
df_active_batters = df_active_batters[df_active_batters['finalGame'] > pd.to_datetime('2015-01-01')]
```

```python
df_active_batters['debut'] = df_active_batters['debut'].dt.date
```

```python
df_active_batters['debut'] = df_active_batters['debut'].dt.date
```

```python
df_active_batters_by_year = df_active_batters
```

```python
goldy_stats = df_active_batters[df_active_batters['playerID'] == 'goldspa01']
```

```python
goldy_stats.head().T
```

```python
rows_to_drop = ['lgID', 'birthYear', 'birthMonth', 'birthDay', 'birthCountry', 'birthState', 'birthCity',
                'deathYear', 'deathMonth', 'deathDay', 'deathCountry', 'deathState', 'deathCity',
                'nameGiven', 'weight', 'height', 'bats', 'throws', 'retroID', 'bbrefID']
goldy_stats.drop(rows_to_drop, inplace=True, axis=1)
```

```python
goldy_stats
```

```python
goldy_stats['name'] = goldy_stats['nameFirst'] + ' ' + goldy_stats['nameLast']
```

```python
goldy_stats.drop(['nameFirst', 'nameLast'], inplace=True, axis=1)
```

```python
aggregations = {'yearID': {'Years': lambda x: max(x) - min(x) + 1},
                'G': {'Games': 'sum'},
                'AB': {'ABs': 'sum'},
                'R': {'Runs': 'sum'},
                'H': {'Hits': 'sum'},
                '2B': {'Doubles': 'sum'},
                '3B': {'Triples': 'sum'},
                'HR': {'HRs': 'sum'},
                'RBI': {'RBIs': 'sum'},
                'SB': {'SBs': 'sum'},
                'CS': {'CS': 'sum'},
                'BB': {'Walks': 'sum'},
                'SO': {'SOs': 'sum'},
                'IBB': {'IBBs': 'sum'},
                'HBP': {'HsBP': 'sum'},
                'SH': {'SacBunts': 'sum'},
                'SF': {'SacFlies': 'sum'},
                'GIDP': {'GIDP': 'sum'}
}
goldy_totals = goldy_stats.groupby(['playerID', 'debut', 'finalGame', 'name']).agg(aggregations)
```
