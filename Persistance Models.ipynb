```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('src/data/condensed.csv')
df
```

```python
avg_cols = [col for col in df.columns if 'avg' in col]
```

```python
df = df[avg_cols]
df
```

```python
y_true = df.pop('year7_avg')
```

```python
def get_metrics(style, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = mse ** 0.5
    score = r2_score(y_true, y_pred)
    print(style + '\n'\
          'MSE:\t{:0.7f}\n'\
          'RMSE:\t{:0.7f}\n'\
          'R2:\t{:0.3f}'.format(mse, rmse, score))
```

```python
y_pred = df['year6_avg']
```

```python
get_metrics('Carry Over', y_true, y_pred)
```

```python
y_pred = df.mean(axis=1)
```

```python
get_metrics('Total Average', y_true, y_pred)
```
